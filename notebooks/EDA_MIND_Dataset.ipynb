This notebook covers the initial analysis of the MIND dataset to prepare for model training and simulation.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cell 1: Imports and Loading Data
news_cols = ['NewsID', 'Category', 'SubCategory', 'Title', 'Abstract', 'URL', 'TitleEntities', 'AbstractEntities']
news = pd.read_csv('../data/news.tsv', sep='\t', names=news_cols)

behav_cols = ['ImpressionID', 'UserID', 'Time', 'History', 'Impressions']
behaviors = pd.read_csv('../data/behaviors.tsv', sep='\t', names=behav_cols)

print(f"Total News Articles: {len(news)}")
print(f"Total User Impressions: {len(behaviors)}")
```

```python
# Cell 2: Data Cleaning
news['Abstract'] = news['Abstract'].fillna(news['Title'])
behaviors['Time'] = pd.to_datetime(behaviors['Time'])
```

```python
# Cell 3: Click-Through Rate (CTR) Analysis
all_impressions = behaviors['Impressions'].str.split().explode()
clicks = all_impressions.apply(lambda x: int(x.split('-')[1]))

ctr = clicks.mean()
print(f"Average Click-Through Rate: {ctr:.2%}")

sns.countplot(x=clicks)
plt.title('Distribution of Clicks (1) vs Skips (0)')
plt.show()
```

```python
# Cell 4: News Category Distribution
plt.figure(figsize=(12,6))
news['Category'].value_counts().plot(kind='bar')
plt.title('News Articles by Category')
plt.ylabel('Count')
plt.show()
```

```python
# Cell 5: Text Length (For Justin’s CNN)
news['TitleLength'] = news['Title'].apply(lambda x: len(str(x).split()))
print(f"Average Title Length: {news['TitleLength'].mean():.2f} words")
print(f"Max Title Length: {news['TitleLength'].max()} words")

sns.histplot(news['TitleLength'], bins=20)
plt.title('Distribution of News Title Lengths')
plt.show()
```
